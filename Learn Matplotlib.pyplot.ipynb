```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

x = np.array([10,120])
y = np.array([12,144])

plt.plot(x,y)
plt.show()
```

```python
# If you just want the points then
plt.plot(x,y,'o')
plt.show()
```

```python
# If we do not send the points for x co-ordinate they are 0 1 2 3 by default
plt.plot(y)
plt.show()
```

```python
# To highlight the points use marker 
plt.plot(x,y,marker='o')
plt.show()
```

### Format Strings fmt

#### You can use also use the shortcut string notation parameter to specify the marker.

#### This parameter is also called fmt, and is written with this syntax:
marker|line|color

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r') 
# 'o' means we want round markers; ':' means we want dotted line; 'r' means we want red color
plt.show()
```

```python
#Use xlabel and ylabel to add labels to the plot
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x,y,'*-b')

plt.xlabel("Average pulse")
plt.ylabel("Calorie Burnage")
plt.title("Pulse vs Calories")
#Adding grid will create a grid in the graph
plt.grid()
#Grid can also be only x axis or only y axis with different linestyles and color and linewidth
plt.show()
```

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
# This means we have one row and two columns out of which this is the first sub-plot
plt.plot(x,y)
plt.title("First subplot")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
# This means of the 1,2 figure this sub-plot is the second one
plt.plot(x,y)
plt.title("Second subplot")

#This is the main title
plt.suptitle("Main title")
plt.show()
```

```python
# Using colors for points and color maps 

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x,y, c=colors, cmap='viridis')
plt.colorbar()
plt.show()
```

```python
# Plotting bar graph and horizontal bar graph 
x = np.array(["A", "B", "C", "D", "E", "F"])
y = np.array([3, 8, 1, 10, 6, 7])

plt.bar(x,y, color="hotpink")
plt.title("Bar graph")
plt.show()

plt.barh(x,y, color="#688022", height=0.7)
plt.title("Horizontal Bar Graph")
```
